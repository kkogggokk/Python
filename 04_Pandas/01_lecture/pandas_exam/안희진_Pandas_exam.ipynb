```python
# 2.  어떤 컬럼들이 범주형 컬럼인지 확인.
# 변수형 컬럼은 어떻게 알수 있는거지? 
forest_df.info() # TODO 
```

```python
# 3. 월(month)과 요일(day) 의 고유한(unique) 값들과 각각의 행수 조회
forest_df['month'].nunique() # 12
forest_df['day'].nunique() # 7
```

```python
# 4. 요일(day) 별 wind의 평균을 조회
forest_df.groupby('day')['wind'].mean()
```

```python
# 5. 요일(day) 별 wind, rain의 평균을 조회
forest_df.groupby('day')[['wind','rain']].mean()
```

```python
# 6. 요일(day) 별 wind, rain의 평균과 합계과 개수 조회
forest_df.groupby('day')[['wind','rain']].agg(['mean','count'])
```

```python
# 7. 요일(day) 별로 wind는 평균을 rain은 합계와 최대값을 조회
forest_df.groupby('day').agg({'wind' : 'mean',
                             'rain' : ['sum','max']})
```

```python
# 8. 월(month) 그리고 요일별(day) temp의 최대값과 최소값을 조회 --> 복수열 기준 그루핑 
forest_df.groupby(['month','day'])['temp'].agg(['max','min'])
```
