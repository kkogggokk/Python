3. 배열의 합을 구한다.
4. 합의 값이 maxweight보다 작거나 같아야 한다. 
5. 이상이 되면 앞의 내용을 pop, 삭제한다. count+1 한다.
6. 반복한다. 

```python
weights = [1,1,1,7,7,7]
maxWeight = 8

lst = list()
count = 0

for index, value in enumerate(weights) : 
    print("index, value : ",index, value)
    if sum(lst) <= maxWeight : 
        lst.append(value) 
        print("lst : ", lst)
    else : 
        count += 1 
        del lst[:index-1]
        print("lst : ",lst)
        
    print(count)
```

```python
class Solution:
    def solution(self, weights, maxWeight):
        lst = list()
        count = 0

        while True :

            if weights != [] :
                count +=1 

            for index, value in enumerate(weights) :
#                 print("--------",lst, count)
                lst.append(value)

                if sum(lst) <= maxWeight : 
                    continue 
                else : 
                    del lst[:len(lst)-1]
                    count += 1

            break
        return count


weights1 = [1,1,1,7,7,7]
maxWeight1 = 8
test = Solution()
print(test.solution(weights1,maxWeight1))
```
