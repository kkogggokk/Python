```python
class Solution:
    def solution(self, weights, maxWeight):
        lst = list()
        count = 0

        while True :

            if weights != [] :
                count +=1 

            for index, value in enumerate(weights) :
#                 print("--------",lst, count)
                lst.append(value)

                if sum(lst) <= maxWeight : 
                    continue 
                else : 
                    del lst[:len(lst)-1]
                    count += 1

            break
        return count


weights1 = [1,1,1,7,7,7]
maxWeight1 = 8
test = Solution()
print(test.solution(weights1,maxWeight1))
```
