## 19.1 중첩루프 사용하기 
- 반복문의 루프 인덱스 변수는 index의 첫글자를 따서 i를 사용
- 반복문이 두개 이므로 i, j 사용 
- i를 사용하는 바깥쪽 루프 : 세로 방향 처리 
- j를 사용하는 안쪽 루프 : 가로 방향 처리 

- 중첩루프는 2차원 평면을 다룰 수 있으므로 이미지처리, 영상처리, 좌표계 처리 등에 주로 사용

```python
for i in range(5) :
    for j in range(3) : 
        print('j:',j,sep='', end=' ') # 안쪽 루프 : 가로 처리 방향
    print('i:',i,'\\n',sep='') # 바깥루프 : 세로 처리 
```

## 19.2 사각형으로 별 출력하기 

```python
# 5*5 사각형 형태로 별 그리기 
for i in range(5) :
    for j in range(5) : 
        print('*',end=' ')
    print()
```

### 19.2.1 사각형 모양 바꾸기 

```python
for i in range(3) : #세로방향 3
    for j in range(7) :  #가로방향 7개
        print('*',end='')
    print()
```

## 19.3 계단식으로 별 출력하기

```python
for i in range(5) : 
    for j in range(5) : 
        if i >= j : 
            print('*', end='')
    print()
```

### 19.3.1 대각선으로 별 출력하기 

```python
for i in range(5) :
    for j in range(5) :
        if i == j :
            print('*',end='')
    print()
#공백 처리가 되지 않았다. 
```

```python
for i in range(5) :
    for j in range(5) :
        if i == j :
            print('*',end='')
        elif i > j : 
            print(' ', end='')
    print()
```
