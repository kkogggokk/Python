## 17.1 while반목문 사용하기 
- 조건식이 참이면 코드를 계속반복하고, 거짓이면 반복문을 끝낸 뒤 다음 코드를 실행
- 조건식 → 반복할 코드 및 변화식 → 조건식 으로 순환하는 부분이 루프 
```python
초기식
while 조건식 : 
    반복할 코드
    변화식 
```

````python
i = 0
while i < 100 :
    print('Hello,world')
    i += 1 
````

````python
### 17.1.1 초깃값을 1부터 시작하기 
````

````python
i = 1 
while i <= 100 :
    print('Hello, world', i)
    i += 1 
````

````python
### 17.1.2 초깃값을 감소 시키기 
````

````python
i = 100 
while i > 0 : 
    print('Hello, world', i )
    i -= 1 
````

````python
###  17.1.3 입력한 횟수대로 반복하기 
count = int(input())
i = 0 

while i < count :
    print('hello, world',i)
    i += 1 
````

````python
count = int(input())

while count > 0 : 
    print('hello, world',count)
    count -= 1 
````
