## 18.6 심사문제 : 두 수 사이의 숫자 중 3으로 끝나지 않는 숫자 출력하기 
- 

```python
start, stop = map(int,input().split())

i = start 
while True : 
    for i in range(start, stop+1) : 
        if i % 10 == 3 :
            i += 1
            continue 
        print(i, end = ' ')
    print(i, end = ' ')
    i += 1 
```

- whiled이 입력 받은 stop이 되면 멈춰야 한다.
- start, stop까지 3으로 끝나는 수를 처리 해야한다.
