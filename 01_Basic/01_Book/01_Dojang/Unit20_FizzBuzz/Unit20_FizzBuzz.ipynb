## 20.1 1부터 100까지 숫자 출력하기 

```python
for i in range(1,101) : 
    print(i)
```

```python
## 20.2 3의 배수일 때와 5의 배수일때 처리하기 
# 3배수일 때 Fizz, 5의 배수일때 Buzz 처리 
# 3과 5의 공배수 FizzBuzz 처리 

for i in range(1,101) : 
    if i % 3 == 0 and i % 5 == 0 :
        print('FizzBuzz')
    elif i % 3 == 0 :
        print('Fizz')
    elif i % 5 == 0 : 
        print('Buzz') 
    else : 
        print(i)
```

## 20.4 논리 연산자를 사용하지 않고 3과 5의 공배수 처리하기 
- 논리연산자 and를 사용하지 않고 3과 5의 공배수를 검사하려면? 

```python
for i in range(1, 101) :
    if i % 15 == 0 :
        print('FizzBuzz')
    elif i % 3 == 0 :
        print('Fizz')
    elif i % 5 == 0 :
        print('Buzz')
    else :
        print(i)
```
