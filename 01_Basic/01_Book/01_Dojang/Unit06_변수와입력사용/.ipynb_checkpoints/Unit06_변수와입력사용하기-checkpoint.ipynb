- 변수를 사용해서 결과를 저장해야 한다. 
- 변수(variabl)를 만들어 결과를 저장해보자. 

## 6.1 변수만들기
- 파이썬에서는 다음 그림과 같은 형식으로 코드를 입력하여 변수를 만든다. 

- 변수이름 = 값 형식
- 변수가 생성되는 동시에 값이 할당(저장)된다. 
- 변수 이름은 다음과 같은 규칙을 지켜야 한다. 
    - 영문 문자와 숫자를 사용할 수 있다. 
    - 대소문자를 구분한다. 
    - 문자부터 시작해야하며 숫자부터 시작하면 안된다. 
    - \_(밑줄)로 시작할 수 있다.
    - 특수문자(+, -, *, /, \$, \@, \&, \% 등)는 사용할 수 없다. 
    - 파이썬의 키워드(if, for, while, and, or 등)는 사용할 수 없다. 

```python
# 숫자 저장(할당)
x = 10
print(x) # 10

# 문자열도 넣을 수 있다. 
y = 'Hello, world!' 
print(y) # Hello, world!
```

### 6.1.1 변수의 자료형 알아내기 
- type(변수) 
- type에 변수를 넣으면 변수(객체)의 자료형이 나온다. 

```python
print(type(x)) # <class 'int'>
print(type(y)) # <class 'str'>
```

- x에는 정수 10이 들어있으므로 int, y에는 문자열이 들어있으므로 str이 나온다. 
- 변수의 자료형은 변수에 들어가는 값에 따라 달라진다. 

### 6.1.2 변수 여러개를 한번에 만들기 
- 변수명1, 변수명2, 변수명 3 = 값1, 값2, 값3 형식으로 변수를 ,(콤마)로 구분한 뒤 각 변수에 할당될 값을 지정해준다. 
- 변수와 값의 개수는 동일하게 맞춰줘야 하며 나열된 순서대로 값이 할당된다. 
- 만약 변수와 값의 개수가 맞지 않으면 에러가 발생한다. 

```python
x, y , z = 10, 20, 30 
print(x, y, z) # 10 20 30

a = b = c = 10 
print(a, b, c) # 10 10 10

n1, n2, n3 = 1, 2 # error : 변수와 값의 개수가 맞아야 한다. 
```

```python
# 응용 : 두 변수의 값 바꾸기 
x, y = 10, 20
x, y = y, x 
print(x, y) # 20 10 
```

## 6.2 변수로 계산하기 
- 변수는 변수끼리 계산할 수 있고, 계산 결과를 다른 변수에 할당할 수 있다. 

### 6.2.1 산술 연산 후 할당 연산자 사용하기 
- 변수 한개에서 값의 변화를 계속 유지하려면 계산 결과를 변수에 저장해야 한다. 
- 산술 연산자 앞에 =(할당 연산자)를 붙이면 연산 결과를 변수에 저장한다. (+ =처럼 두 연산자를 공백으로 띄우면 안 된다.)
- 산술 연산 후 할당 연산자를 사용할 때 주의할 점 : 만들지 않은 변수에 할당하면 에러가 발생 
- 왜냐하면 변수가 없어 산술연산 후 변수가 없어서 할당이 안되기 때문이다. 
- 따라서 연산 후 할당을 하려면 값이 들어 있는 변수를 사용해야 한다. 

```python
a = 10
b = 20
c = a + b 
print(c) # 30

# 변수 한개에서 값의 변화를 계속 유지하려면 계산 결과를 다시 변수에 저장해야 한다. 
a = 10
a + 20 # 30 
a  # 10 
a = a + 20 # 30 

# a 를 두번 입력해야하는 번거로움 
a = 10 
a += 20 # a와 20을 더한 후 결과를 다시 a 에 저장 
a # 30 

# 변수 선언 후 할당 
d += 10 # NameError: name 'd' is not defined
```

> 참고 > 부호붙이기 
> - 부호를 붙여야 하는 경우 값이나 변수 앞에 양수, 음수 부호를 붙이면 된다. 

```python
x = -10
print(+x) # -10 
print(-x) # 10
```

### 6.3.3 두 숫자의 합 구하기 

```python
num1 = input('첫번째 숫자를 입력하세요 :') # 10
num2 = input('두번째 숫자를 입력하세요 :') # 20 

print(num1 + num2) # 1020
print(type(num1)) # <class 'str'>
```

- 10 + 20 은 30이 나와야 하는데 1020으로 출력되었다.
