#### 1.5.2.1 figure.add_subplot
- figure객체에 axes를 추가하는 형태이다.
- figure.add_subplot(nrows, ncols, index)
    - nrows : 총 행수
    - ncols : 총 열수
    - index : axes위치
    
# TODO : 아래 그림에 대한 설명 추가 


```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (10, 10)) # figure객체 생성

#figure에 axes(subplot) 을 추가 : nrows(총행수), ncols(총열수), index(axes위치) 지정
ax1 = fig.add_subplot(2,2,1) 
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(223) # 쉼표없이 지정가능
ax4 = fig.add_subplot(224)

# 그래프 내부에 글씨를 써보자 : ax1.text(0.5,0.5,"안녕") 
axes = [ax1, ax2, ax3, ax4]
for i, ax in enumerate(axes, start = 1):
    txt = f'ax {i}'
    ax.text(0.5, 0.5, txt, fontsize = 25) # text(x위치, y위치, 텍스트) : 텍스트의 left botton이 위치할 곳을 지정

plt.show()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (10,10))

ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 3)
ax3 = fig.add_subplot(1, 2, 2) 

for i, ax in enumerate([ax1, ax2, ax3], start = 1):
    ax.text(0.5, 0.5, f'ax{i}', fontsize = 20)
    ax.set_title(f'ax label{i}')
    
    # 각 축에 레이블 붙이기
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    
plt.tight_layout() # 각각의 그래프 사이 간격주기 
plt.show()
```

## 2.1 꺽은선그래프 plot()
- 점과 점을 선으로 연결한 선(꺾은선)그래프를 표현한다.
- 시간의 흐름에 따른 변화를 표현할 때 많이 사용한다. 시계열 
- plt.plot([x], y)
    - x, y의 인수는 {리스트, 튜플, 넘파이의 ndarray, 판다스의 시리즈}의 객체들을 넣을 수 있다. 
    - x와 y의 사이즈는 같아야 한다. 
    - x값은 생략할 수 있다. 
    - 인수가 하나인 경우 y축의 값으로 설정되고 x값은 0에서 len(y)-1범위로 지정된다. 
- 하나의 axes(subplot)에 여러개의 선 그리려면? 같은 axes에 plot()를 여러번 실행한다. 
- 선스타일 지정 가능하다. [line-style reference](https://matplotlib.org/3.0.3/gallery/lines_bars_and_markers/line_styles_reference.html)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(1, 10, num = 100) # 1 - 10

plt.figure(figsize = (5,5))
plt.plot(x, x + 3, linestyle = '-', linewidth = 5) # 선 긁기 지정 가능
plt.plot(x, x + 2, linestyle = '--')
plt.plot(x, x + 1, linestyle = ':')
plt.plot(x, x, linestyle = '-.')

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data/서울시 연도별 황사 경보발령 현황.csv')
df.head()
```

```python
# 최대농도(㎍/㎥/시) 컬럼 조회 시 불편하니 이름을 '최대농도'로 변경하자
df.rename({df.columns[-1] : '최대농도'}, axis = 1, inplace = True)
```

```python
# 연도별 관측일수의 변화 흐름을 시각화 해보자. 
plt.figure(figsize = (10,5))

# x : 시간(년도), y : 관측일수
plt.plot(df['년도'], df['관측일수'])

# 제목, 각 축의 레이블을 입력하자
plt.title('연도별 황사 관측일수 변화')
plt.xlabel('년도') 
plt.ylabel('관측일수')


plt.show()
```

```python
# 연도별 주인보 발령횟수와 경보 발령횟수의 변화를 그래프로 시각화 해보자 
plt.figure(figsize = (10, 5))

# plt 설정
plt.plot(df['년도'], df['주의보 발령횟수'], label = '주의보')
plt.plot(df['년도'], df['경보 발령횟수'], label = '주의보', alpha = 0.7) 

# 타이블, 레이블 설정
plt.title('연도별 주의보 및 경보 발령횟수의 변화')
plt.xlabel('년도')
plt.ylabel('발령횟수')

# ticks 설정 
plt.xticks(df['년도'], rotation = 45)
plt.yticks(range(0, 5))
# plt.yticks(df['주의보 발령횟수'].unique())

plt.grid()
plt.legend(fontsize = 15, fancybox = True, shadow = True, ncol = 2)
plt.show()
```

- 두 개의 선을 구분하기 위해서 label로 구분한다. 
- 선이 겹치는 것을 구분하기 위해 alpha(투명도)값을 0(투명) - 1(불투명) 지정한다.
- yticks 설정 전에는 0.5 단위도 표기가 되었다. 년도로 구분하기 때문에 range를 써서 1년 단위로 표현했다. 혹은 df['주의보 발령횟수'].unique()로도 설정이 가능하다
- legend의 ncol

## 2.2 산점도 scatter()

- **변수간의 상관성**이나 관측값들 간의 군집 분류를 확인할 수 있다. 
- scatter(x, y) 메소드를 사용한다.
    - x, y값들은 모두 매개변수로 전달된다.
    - x, y의 인수는 스칼라 실수나 리스트형태(리슽, 튜플, 넘파이의 ndarray, 판다스의 Series)의 객체들을 넣는다.
    - x, y 원소의 수는 같아야 한다. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.arange(1, 1000, 50)
y = np.arange(1, 1000, 50)
z = np.arange(1, 1000, 20)

plt.scatter(x, y)
plt.show()
```

## 2.3 막대그래프 Bar()
- 수량을 막대형식으로 나타낸 그래프 
- axes.bar(x, height) 메소드
    - 수직막대 그래프 
    - x : 분류 값
    - height : 막대높이, 갯수
- axes.harh(y, width) 메소드 
    - 수평막대 그래프 
    - y : 
    - wideh : 막대 너비

```python
import numpy as np
import matplotlib.pyplot as plt

fruits = ['배', '사과', '수박']
amount = [10,15,5]
fig, axes = plt.subplots(2,1, figsize=(5, 8))

axes[0].bar(fruits, amount, width=0.5) #0 ~ 1 (기본: 0.8)
axes[0].set_title('수직막대그래프')
axes[0].set_xlabel('과일')
axes[0].set_ylabel('수량')
axes[0].grid(True)


axes[1].barh(fruits, amount, height=0.4)
axes[1].set_title('수평막대그래프')
axes[1].set_ylabel('과일')
axes[1].set_xlabel('수량')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

- pie(x, labels, autopct, textprops, explode) 메소드
    - x : 
    - labels : 
    - autopct : 
    - textprops :
    - explode 

```python
import matplotlib.pyplot as plt

labels = ['사과', '배', '복숭아', '수박', "딸기"]
amount = [20,10,40,5,60]

plt.figure(figsize=(10,10))
plt.pie(amount, labels=labels, autopct="%.2f%%", 
       shadow=True,
       textprops={'fontsize':15},
       explode=[0,0.2,0,0,0.2]
       )

plt.show()
```
